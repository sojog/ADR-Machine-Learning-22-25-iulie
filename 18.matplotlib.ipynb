```python
x = np.arange(10)
plt.plot(x, x);
plt.scatter(x, x, c="#32FF13");
```

```python
x = np.random.randint(1, 10, 5)
plt.plot(x, x);
plt.scatter(x, x, c="#32FF13");
```

```python
x = np.random.randint(1, 10, 5)
y = np.random.randint(1, 10, 5)
print("x=", x, "y=", y)
plt.plot(x, y);
plt.scatter(x, y, c="#32FF13");
```

```python
x = np.arange(10)
y = x ** 2
plt.plot(x, y);
plt.scatter(x, y, c="#32FF13");
```

```python
x = np.arange(10)
y = x * 2
plt.plot(x, y);
plt.scatter(x, y, c="#32FF13");
```

```python
x = [2, 3, 4, 5, 6]
y = [4, 7, 8, 9, 11]
plt.plot(x, y);
plt.scatter(x, y, c="#32FF13");
```

```python
x = [2, 3, 4, 5, 6]
y = [4, 7, 8, 9, 11]
y_2 = [4, 6, 8, 10, 12]
plt.plot(x, y);
plt.scatter(x, y, c="#32FF13");
plt.plot(x, y_2);
```

```python

def functie_liniara(x):
    return 1.5 * x + 10

```

```python
x = [2, 4, 5, 7, 10]
y = list(map(functie_liniara, x))
plt.scatter(x, y);
```
