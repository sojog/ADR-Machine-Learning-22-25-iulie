```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = LinearRegression()
model
```

```python
apartamente = np.array([(37, 65_000), (55, 80_000), (80, 120_000), (60, 95_000), (92, 205_000), (41, 75_000), (170, 400_000)])
apartamente[apartamente[:, 0].argsort()]

suprafete = apartamente[:, 0] 
preturi  = apartamente[:, 1] / 1000
plt.scatter(suprafete, preturi)
```

```python
suprafete, preturi
```

```python
suprafete = suprafete.reshape(-1, 1)
```

```python
model.fit(suprafete, preturi)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
(np.float64(2.594026044986795), np.float64(-49.68627629541933))
```

```python
model.predict([[100]])
```

```python
model.predict([[100], [75]])
```

```python
   
def functie_regresie(x):
    return model.coef_ * x + model.intercept_

plt.scatter (suprafete,preturi) 
plt.plot (suprafete,list(map(functie_regresie, suprafete)) )
```
