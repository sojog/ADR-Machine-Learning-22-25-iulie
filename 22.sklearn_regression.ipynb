```python
x, y = make_regression(n_samples=100, n_features=1)
plt.scatter(x, y);
```

```python
x[20]
```

```python
x[20] = 0.9
```

```python
plt.scatter(x, y);
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x, y)
```

```python
plt.scatter(x, y);
plt.plot(x, list(map(lambda x: model.coef_ * x + model.intercept_, x)));
```
