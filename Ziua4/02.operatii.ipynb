# Kernel

```python
blurred_image1 = cv2.blur(normal_image, (3,3))
plt.imshow(blurred_image1);
```

```python
blurred_image1 = cv2.blur(normal_image, (30,30))
plt.imshow(blurred_image1);
```

```python
blurred_image3 = cv2.blur(normal_image, (100,100))
plt.imshow(blurred_image3);
```

```python
left_edge_kernel = np.array([[-1, 0, 1], [-1, 0, 1],[-1, 0, 1]] )
left_edge_kernel
```

```python
left_kernel_image = cv2.filter2D(normal_image, -1, left_edge_kernel)
plt.imshow(left_kernel_image)
```

```python
top_kernel = np.array([[-1, -1, -1], [0, 0, 0],[1, 1, 1]] )

top_kernel_image = cv2.filter2D(normal_image, -1, top_kernel)
plt.imshow(top_kernel_image)
```

```python
top_kernel = np.array([[-3, -3, -3], [0, 0, 0],[3, 3, 3]] )

top_kernel_image = cv2.filter2D(normal_image, -1, top_kernel)
plt.imshow(top_kernel_image)
```
